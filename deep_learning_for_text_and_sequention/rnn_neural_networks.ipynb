```python
file_name = "C:\\Users\\1\\Desktop\\reinforcment_deep_ML\\jena_climate_2009_2016.csv"
data_file = open(file_name, "r")
data = data_file.read()
data_file.close()

lines = data.split("\n")
headers = lines[0].split(",")
lines = lines[1:]

print(f"DATA BASE HEADERS: {headers}")
data_tensor = np.zeros(shape=(len(lines), len(headers) - 1))
for (line_index, line) in enumerate(lines):
    values = [float(value) for value in line.split(",")[1:]]
    data_tensor[line_index, :] = values

print(f"[DATA TENSOR SHAPE: {data_tensor.shape}]")
```

```python
shown_features = 8

plt.style.use("dark_background")
fig, axis = plt.subplots(nrows=shown_features)
colors = ["red", "blue", "green", "y", "coral", "lightseagreen"]

for feature in range(shown_features):
    axis[feature].plot(range(1, data_tensor[:10000].shape[0] + 1), data_tensor[:10000, feature], color=rd.choice(colors), label=headers[feature])
    axis[feature].legend(loc="upper left")

plt.show()
```

```python
data_tensor_std = data_tensor[:200000] - data_tensor[:200000].mean() / (data_tensor[:200000] - data_tensor.mean()) ** 2
```

```python
def data_generator(data, lookback, delay, min_index, max_index,
                   shuffle=False, batch_size=128, step=6):
    if max_index is None:
        max_index = len(data_tensor_std) - delay - 1
    index = min_index + lookback

    while True:

        if shuffle:
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size)
        else:
            if index + batch_size >= max_index:
                index = min_index + lookback
            rows = np.arange(index, min(index + batch_size, max_index))
            index += len(rows)
        
        samples = np.zeros(shape=(
            len(rows),
            lookback // step,
            data.shape[-1]
        ))
        targets = np.zeros((len(rows), ))
        for (row_index, row) in enumerate(rows):
            indices = range(rows[row_index] - lookback, rows[row_index], step)
            samples[row_index] = data[indices]
            targets[row_index] = data[rows[row_index] + delay][1]
        
        yield samples, targets
```

```python
lookback = 1440
step = 6
delay = 144
batch_size = 128

train_gen = data_generator(data_tensor,
                           lookback=lookback,
                           delay=delay,
                           min_index=0,
                           max_index=200000,
                           shuffle=True,
                           step=step,
                           batch_size=batch_size)

val_gen = data_generator(data_tensor,
                         lookback=lookback,
                         delay=delay,
                         min_index=200001,
                         max_index=300000,
                         step=step)

test_gen = data_generator(data_tensor,
                          lookback=lookback,
                          delay=delay,
                          min_index=300001,
                          max_index=None,
                          step=step,
                          batch_size=batch_size)

val_steps = (300000 - 200001 - lookback) // batch_size
test_steps = (len(data_tensor) - 300001 - lookback) // batch_size

```
